```python
import zipfile
import xml.etree.ElementTree as ET

zip_path = r"C:\Users\zengo\OneDrive\Magic Briefcase\Article Ideas\Nonprofit Curriculum Development Strategies\IRS XML Zip Files\IRS XML Zip Files 2024\2024_TEOS_XML_01A.zip"

def detect_form_type(root):
    for elem in root.iter():
        tag = elem.tag.split("}")[-1]
        if tag.startswith("IRS990"):
            return tag
    return "Unknown"

form_990_files = []

with zipfile.ZipFile(zip_path, 'r') as z:
    xml_files = [f for f in z.namelist() if f.endswith(".xml")]

    for xml_file in xml_files:
        with z.open(xml_file) as file:
            tree = ET.parse(file)
            root = tree.getroot()
            form_type = detect_form_type(root)

            if form_type == "IRS990":
                form_990_files.append(xml_file)

print("Total IRS990 filings found:", len(form_990_files))
print("Example files:", form_990_files[:5])
```

```python
# ============================================================
# INSPECT ALL XML FIELDS INSIDE A TRUE IRS990
# ============================================================

import zipfile
import xml.etree.ElementTree as ET

zip_path = r"C:\Users\zengo\OneDrive\Magic Briefcase\Article Ideas\Nonprofit Curriculum Development Strategies\IRS XML Zip Files\IRS XML Zip Files 2024\2024_TEOS_XML_01A.zip"

# Use one confirmed IRS990 file from your list
sample_file = "2024_TEOS_XML_01A/202303569349300000_public.xml"

def strip_ns(tag):
    return tag.split("}")[-1]

with zipfile.ZipFile(zip_path, 'r') as z:
    with z.open(sample_file) as file:
        tree = ET.parse(file)
        root = tree.getroot()
        
        # Find IRS990 section specifically
        irs990_section = None
        for elem in root.iter():
            if strip_ns(elem.tag) == "IRS990":
                irs990_section = elem
                break
        
        if irs990_section is None:
            print("IRS990 section not found.")
        else:
            tags = sorted(set(strip_ns(e.tag) for e in irs990_section.iter()))
            
            print("Total unique IRS990 fields:", len(tags))
            print("\nFirst 50 fields:\n")
            for tag in tags[:150]:
                print(tag)
```
